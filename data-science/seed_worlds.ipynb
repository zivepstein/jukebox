```python
seed_set = pd.read_csv("seed_set.csv")
```

```python
add_feedback = """ INSERT INTO gg_feedback_prod VALUES (now(), %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""
add_engagement = """ INSERT INTO gg_engagement_prod VALUES (now(), %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

def write2database(conn, table, data):
	cur = conn.cursor()
	cur.execute(table, (data))
	conn.commit()
	cur.close()
```

```python
ip = -1
condition = -1
truncation = 0.05
noise_seed = 20
video = -1
meta = "worldseed"

worlds = range(673,689)
```

```python
engagement_raw = []
feedback_raw = []
for world in worlds:
    for i,r in seed_set[seed_set['is_seed']==1].iterrows():
        if (r['is_fed']==0):
            url = r['url']
            species = "{}-{}".format(int(r['cat_a']),int(r['cat_b']))
            name = r['name']
            data = (ip, condition, url, species, truncation, noise_seed, name, -1, -1, -1, -1, world, video, meta)
            print(data)
            write2database(conn, add_feedback, data)
            feedback_raw.append(data)
        else:
            url = r['url']
            species = "{}-{}".format(int(r['cat_a']),int(r['cat_b']))
            food = r['feed_total']
            name = r['name']
            engagement_data = (ip, condition, url, species, truncation, noise_seed, name, 0, food, world, video, meta)
            print(engagement_data)
            write2database(conn, add_engagement, engagement_data)
            engagement_raw.append(engagement_data)
```
